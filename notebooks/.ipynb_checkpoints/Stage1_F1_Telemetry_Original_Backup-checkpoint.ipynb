# F1 Telemetry Analysis - Stage 1 (Easy)
**Objective:** Analyze the fastest lap of drivers from the 2023 Bahrain GP Qualifying session, visualize speed, throttle, and brake data.

```python
# Libraries
import fastf1
import matplotlib.pyplot as plt
import os

# Enable caching
cache_path = "cache"
if not os.path.exists(cache_path):
    os.makedirs(cache_path)
fastf1.Cache.enable_cache(cache_path)
```

```python
# Print all drivers (number - abbreviation - last name)
for drv in session.drivers:
    info = session.get_driver(drv)
    print(f"{info['DriverNumber']} - {info['Abbreviation']} ({info['LastName']})")
```

```python
# Pick fastest laps
ver_lap = session.laps.pick_drivers("VER").pick_fastest()
lec_lap = session.laps.pick_drivers("LEC").pick_fastest()
```

```python
# Get telemetry data
ver_tel = ver_lap.get_car_data().add_distance()
lec_tel = lec_lap.get_car_data().add_distance()

# Plot speed, throttle, brake
plt.figure(figsize=(12,6))
plt.plot(ver_tel['Distance'], ver_tel['Speed'], label='VER Speed')
plt.plot(ver_tel['Distance'], ver_tel['Throttle'], label='VER Throttle')
plt.plot(ver_tel['Distance'], ver_tel['Brake'], label='VER Brake')
plt.xlabel("Distance (m)")
plt.ylabel("Speed / Throttle / Brake")
plt.title("VER Telemetry - Bahrain 2023 Quali")
plt.legend()
plt.show()
```

```python
# Create plots folder if not exists
if not os.path.exists('plots'):
    os.makedirs('plots')

# Save figure
plt.savefig('plots/VER_Bahrain2023.png', dpi=300)
```

```python
# ===============================
# Stage 1: F1 Telemetry Analysis
# ===============================

# 1️⃣ Imports
import fastf1
import matplotlib.pyplot as plt
import os

# 2️⃣ Create necessary folders
folders = ['cache', 'plots']
for f in folders:
    if not os.path.exists(f):
        os.makedirs(f)
print("Folders ready:", folders)

# 3️⃣ Enable FastF1 cache
fastf1.Cache.enable_cache('cache')

# 4️⃣ Load session
session = fastf1.get_session(2023, 'Bahrain', 'Q')
session.load()
print("Session loaded ✅")

# 5️⃣ Map drivers
print("Drivers in session:")
for drv in session.drivers:
    info = session.get_driver(drv)
    print(f"{info['DriverNumber']} - {info['Abbreviation']} ({info['LastName']})")

# 6️⃣ Pick fastest laps
ver_lap = session.laps.pick_drivers("VER").pick_fastest()
lec_lap = session.laps.pick_drivers("LEC").pick_fastest()
print("Fastest laps selected ✅")

# 7️⃣ Get telemetry
ver_tel = ver_lap.get_car_data().add_distance()
lec_tel = lec_lap.get_car_data().add_distance()

# 8️⃣ Plot telemetry
plt.figure(figsize=(12,6))
plt.plot(ver_tel['Distance'], ver_tel['Speed'], label='VER Speed')
plt.plot(ver_tel['Distance'], ver_tel['Throttle'], label='VER Throttle')
plt.plot(ver_tel['Distance'], ver_tel['Brake'], label='VER Brake')
plt.xlabel("Distance (m)")
plt.ylabel("Speed / Throttle / Brake")
plt.title("VER Telemetry - Bahrain 2023 Quali")
plt.legend()
plt.show()

# 9️⃣ Save plot
plt.savefig('plots/VER_Bahrain2023.png', dpi=300)
print("Plot saved in 'plots/' folder ✅")
```
